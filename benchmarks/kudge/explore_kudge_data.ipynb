```python
import re
from typing import Any, Dict

ANSWER_RE = re.compile(r"\[ANSWER\]\s*([A-Da-d])\s*(?:\[END\])?", re.IGNORECASE)
SECTION_RE = re.compile(
    r"###\s*([^:\n]+)\s*:\s*(.*?)(?=\n?\s*###\s*[^:\n]+\s*:|$)",
    re.DOTALL,
)
OPTION_RE = re.compile(
    r"(?:^|\n|\s)(?:[-*Рђб]\s*)?\(?([A-Da-d])\)?\s*[\).:№╝џсђЂ]\s*",
    re.IGNORECASE,
)

SECTION_ALIASES = {
    "ВДѕвгИ": "question",
    "вгИВаю": "question",
    "ВёаьЃЮВДђ": "choices",
    "в│┤ЖИ░": "choices",
    "ВўхВЁў": "choices",
    "ВёаьЃЮ": "choices",
    "ВёаьЃЮВѓгьЋГ": "choices",
    "вІх": "answer",
    "вІхв│ђ": "answer",
    "ВаЋвІх": "answer",
    "ьњђВЮ┤": "solution",
    "ьЋ┤Вёц": "solution",
}

def _clean_text(value: str) -> str:
    return re.sub(r"\s+", " ", str(value)).strip()

def _section_key(label: str) -> str:
    label = re.sub(r"\s+", "", label.strip())
    return SECTION_ALIASES.get(label, label)

def _parse_options(text: str, *, uppercase_labels: bool = True) -> Dict[str, str]:
    matches = list(OPTION_RE.finditer(text))
    options = {}
    for i, match in enumerate(matches):
        raw_label = match.group(1)
        label = raw_label.upper() if uppercase_labels else raw_label.lower()
        start = match.end()
        end = matches[i + 1].start() if i + 1 < len(matches) else len(text)
        option = re.split(r"\s*#{3,}", text[start:end], maxsplit=1)[0]
        options[label] = _clean_text(option)
    return options

def _strip_inline_options(question: str) -> str:
    first = OPTION_RE.search(question)
    return _clean_text(question[:first.start()]) if first else _clean_text(question)

def parse_mcq(text: str) -> Dict[str, Any]:
    raw_text = str(text).replace("\r\n", "\n").replace("\r", "\n").strip()
    sections = {}
    for label, content in SECTION_RE.findall(raw_text):
        key = _section_key(label)
        sections[key] = content.strip()

    raw_question = sections.get("question", "")
    choices_text = sections.get("choices", "")

    # Fallback for prompts without a recognized choices section. This happens
    # in a few translated rows where options are embedded directly in ВДѕвгИ.
    if raw_question and not choices_text:
        inline_options = _parse_options(raw_question, uppercase_labels=False)
        if inline_options:
            return {
                "question": _strip_inline_options(raw_question),
                "options": {letter.upper(): value for letter, value in inline_options.items()},
                "answer": _clean_text(sections.get("answer", "")),
            }

    # Fallback for prompts without section markers at all.
    if not raw_question:
        raw_question = raw_text
        inline_options = _parse_options(raw_question, uppercase_labels=False)
        if inline_options:
            return {
                "question": _strip_inline_options(raw_question),
                "options": {letter.upper(): value for letter, value in inline_options.items()},
                "answer": "",
            }
        raise ValueError("Could not find question section: ### ВДѕвгИ:")

    if not choices_text:
        labels = ", ".join(sections.keys()) or "none"
        raise ValueError(f"Could not find choices section. Parsed sections: {labels}")

    inline_options = _parse_options(raw_question, uppercase_labels=False)
    options = _parse_options(choices_text, uppercase_labels=True)

    if len(options) < 2:
        raise ValueError(f"Could not parse choices from choices section: {choices_text[:200]!r}")

    answer = _clean_text(sections.get("answer", ""))
    answer_match = ANSWER_RE.search(raw_text)
    if answer_match:
        answer = answer_match.group(1).upper()

    return {
        "question": _strip_inline_options(raw_question),
        "options": options,
        "answer": answer,
    }
```

```python
ANSWER_LETTER_RE = re.compile(r"\[ANSWER\]\s*([A-Da-d])\s*(?:\[END\])?", re.IGNORECASE)
LETTER_PHRASE_RE = re.compile(r"(?:ВаЋвІх|вІх|answer)\s*(?:ВЮђ|віћ|:|№╝џ|=)?\s*\**\(?([A-Da-d])\)?", re.IGNORECASE)
NUMBER_RE = re.compile(r"\d+(?:\.\d+)?")

def _norm_for_match(text):
    text = str(text).lower()
    text = re.sub(r"\s+", "", text)
    text = re.sub(r"[^0-9a-zЖ░ђ-ьъБ╬▒-¤Ѕ╬Љ-╬Есё▒-сЁјсЁЈ-сЁБСИђ-жЙЦ]+", "", text)
    return text

def _numbers(text):
    return set(NUMBER_RE.findall(str(text)))

def _final_answer_region(text):
    text = str(text)
    markers = ["вћ░вЮ╝Вёю", "ЖиИвЪгв»ђвАю", "Вўгв░ћвЦИ вІхв│ђВЮђ", "ВаЋвІх", "answer"]
    starts = [text.lower().rfind(marker.lower()) for marker in markers]
    start = max(starts)
    return text[start:] if start >= 0 else text[-500:]

def extract_answer_letter(chosen, parsed):
    chosen_text = str(chosen)
    final_text = _final_answer_region(chosen_text)

    match = ANSWER_LETTER_RE.search(chosen_text) or LETTER_PHRASE_RE.search(chosen_text)
    if match:
        return match.group(1).upper(), "explicit_letter"

    chosen_norm = _norm_for_match(chosen_text)
    final_norm = _norm_for_match(final_text)
    final_nums = _numbers(final_text)
    scored = []
    for letter, option in parsed["options"].items():
        option_text = str(option)
        option_norm = _norm_for_match(option_text)
        if not option_norm:
            continue

        score = 0.0
        source = "not_found"
        if len(option_norm) >= 3 and option_norm in chosen_norm:
            score = 1000 + len(option_norm)
            source = "matched_option_text"
        else:
            option_nums = _numbers(option_text)
            if option_nums and option_nums.issubset(final_nums):
                score = 100 + len(option_nums)
                source = "matched_option_numbers"
            else:
                tokens = [_norm_for_match(tok) for tok in re.split(r"\s+", option_text)]
                tokens = [tok for tok in tokens if len(tok) >= 2]
                hits = sum(1 for tok in tokens if tok in final_norm)
                if hits >= 2 and hits / max(len(tokens), 1) >= 0.6:
                    score = hits / max(len(tokens), 1)
                    source = "matched_option_tokens"
        scored.append((score, letter, source))

    scored.sort(reverse=True)
    if scored and scored[0][0] > 0 and (len(scored) == 1 or scored[0][0] > scored[1][0]):
        return scored[0][1], scored[0][2]

    return None, "not_found"

def mcq_record(row):
    try:
        parsed = parse_mcq(row["prompt"])
        parse_error = ""
    except Exception as exc:
        parsed = {"question": str(row["prompt"]), "options": {}, "answer": ""}
        parse_error = str(exc)

    answer, answer_source = extract_answer_letter(row["chosen"], parsed) if not parse_error else (None, "parse_error")
    return {
        "id": row["id"],
        "subset": row["subset"],
        "question": parsed["question"],
        "A": parsed["options"].get("A", ""),
        "B": parsed["options"].get("B", ""),
        "C": parsed["options"].get("C", ""),
        "D": parsed["options"].get("D", ""),
        "answer": answer,
        "answer_text": parsed["options"].get(answer, "") if answer else "",
        "answer_source": answer_source,
        "parse_error": parse_error,
        "chosen_model": row.get("chosen_model", ""),
        "rejected_model": row.get("rejected_model", ""),
    }

mcq_df = pd.DataFrame([mcq_record(row) for _, row in challenge_df.iterrows()])
mcq_df.head()
```

## Korean-Hard Manual Labeling Interface

This section filters the challenge data to `Korean-Hard`, parses each item into question/options, shows the original prompt and chosen/reference solution, highlights exact option-text matches in the chosen text, and saves your labels to JSON.


```python
# Widget support for the manual labeling interface.
# Already installed in the project venv, but keep this here if the notebook runs elsewhere.
%pip install ipywidgets
```

```python
from pathlib import Path
import json

LABEL_OUTPUT_PATH = repo_root / "benchmarks" / "kudge" / "kudge_korean_hard_groundtruth_labels.json"

korean_hard_df = challenge_df[challenge_df["subset"] == "Korean-Hard"].reset_index(drop=True)


def _safe_parse_mcq(prompt):
    try:
        parsed = parse_mcq(prompt)
        return parsed, ""
    except Exception as exc:
        return {"question": str(prompt), "options": {}, "answer": ""}, str(exc)


def _existing_records_by_id(path=LABEL_OUTPUT_PATH):
    if not path.exists():
        return {}
    with path.open("r", encoding="utf-8") as fh:
        records = json.load(fh)
    return {str(record.get("id")): record for record in records}


def _existing_groundtruth_by_id(path=LABEL_OUTPUT_PATH):
    return {
        row_id: record.get("groundtruth_answer")
        for row_id, record in _existing_records_by_id(path).items()
        if record.get("groundtruth_answer")
    }


def _row_or_existing(row, key, existing_record, default=""):
    value = row.get(key, default)
    if value is None or str(value) == "nan":
        value = default
    if value == default and existing_record:
        value = existing_record.get(key, default)
    return value


def korean_hard_label_records(labels_by_id=None):
    labels_by_id = labels_by_id or _existing_groundtruth_by_id()
    existing_by_id = _existing_records_by_id()
    records = []
    for _, row in korean_hard_df.iterrows():
        row_id = str(row["id"])
        existing_record = existing_by_id.get(row_id, {})
        prompt = _row_or_existing(row, "prompt", existing_record)
        chosen = _row_or_existing(row, "chosen", existing_record)
        parsed, parse_error = _safe_parse_mcq(prompt)
        if parse_error:
            heuristic_answer, heuristic_source = None, "parse_error"
        else:
            heuristic_answer, heuristic_source = extract_answer_letter(chosen, parsed)
        records.append({
            "id": row_id,
            "subset": _row_or_existing(row, "subset", existing_record),
            "prompt": prompt,
            "chosen": chosen,
            "rejected": _row_or_existing(row, "rejected", existing_record),
            "chosen_model": _row_or_existing(row, "chosen_model", existing_record),
            "rejected_model": _row_or_existing(row, "rejected_model", existing_record),
            "parsed_question": parsed.get("question", ""),
            "parsed_options": parsed.get("options", {}),
            "heuristic_answer": heuristic_answer,
            "heuristic_answer_source": heuristic_source,
            "parse_error": parse_error,
            "groundtruth_answer": labels_by_id.get(row_id),
        })
    return records


def save_korean_hard_labels(labels_by_id=None, path=LABEL_OUTPUT_PATH):
    records = korean_hard_label_records(labels_by_id=labels_by_id)
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open("w", encoding="utf-8") as fh:
        json.dump(records, fh, ensure_ascii=False, indent=2)
    labelled = sum(1 for record in records if record.get("groundtruth_answer"))
    rejected = sum(1 for record in records if str(record.get("rejected", "")).strip())
    print(f"Saved {len(records)} Korean-Hard records to {path}")
    print(f"Labelled: {labelled}/{len(records)} | Rejected present: {rejected}/{len(records)}")
    return records

existing_labels_by_id = _existing_groundtruth_by_id()
korean_hard_records = korean_hard_label_records(existing_labels_by_id)
korean_hard_preview_df = pd.DataFrame(korean_hard_records)
print(f"Korean-Hard items: {len(korean_hard_records)}")
print(f"Existing labels loaded: {len(existing_labels_by_id)}")
print(f"Rejected present: {sum(bool(str(record.get('rejected', '')).strip()) for record in korean_hard_records)}/{len(korean_hard_records)}")
korean_hard_preview_df[["id", "subset", "heuristic_answer", "heuristic_answer_source", "groundtruth_answer", "parse_error"]].head()
```

```python
import html
import re
import ipywidgets as widgets
from IPython.display import HTML, display, clear_output

labels_by_id = dict(existing_labels_by_id)
label_state = {"index": 0}

item_out = widgets.Output()
status_out = widgets.Output()

answer_buttons = [
    widgets.Button(description=letter, button_style="info", layout=widgets.Layout(width="64px", height="36px"))
    for letter in ["A", "B", "C", "D"]
]
unknown_button = widgets.Button(description="UNKNOWN", icon="help-circle", layout=widgets.Layout(width="115px", height="36px"))
clear_button = widgets.Button(description="Clear", icon="trash", layout=widgets.Layout(width="95px", height="36px"))
prev_button = widgets.Button(description="Previous", icon="arrow-left", layout=widgets.Layout(width="115px", height="36px"))
next_button = widgets.Button(description="Next", icon="arrow-right", layout=widgets.Layout(width="105px", height="36px"))
save_button = widgets.Button(description="Save JSON", icon="save", button_style="success", layout=widgets.Layout(width="125px", height="36px"))

jump_box = widgets.BoundedIntText(
    value=1,
    min=1,
    max=max(len(korean_hard_records), 1),
    description="Item",
    layout=widgets.Layout(width="165px"),
)


def _escape(value):
    return html.escape(str(value))


def _current_record():
    label_state["index"] = max(0, min(label_state["index"], len(korean_hard_records) - 1))
    return korean_hard_records[label_state["index"]]


def _highlight_exact_option_matches(text, options):
    escaped_text = _escape(text)
    matches = []
    for letter, option in (options or {}).items():
        option = str(option).strip()
        if len(option) < 2:
            continue
        for match in re.finditer(re.escape(option), str(text), flags=re.IGNORECASE):
            matches.append((match.start(), match.end(), str(letter)))
    if not matches:
        return escaped_text

    matches.sort(key=lambda x: (x[0], -(x[1] - x[0])))
    merged = []
    last_end = -1
    for start, end, letter in matches:
        if start < last_end:
            continue
        merged.append((start, end, letter))
        last_end = end

    raw = str(text)
    parts = []
    cursor = 0
    for start, end, letter in merged:
        parts.append(_escape(raw[cursor:start]))
        parts.append(
            '<mark style="background:#fff59d; padding:1px 3px; border-radius:3px;" '
            f'title="exact match for option {letter}">{_escape(raw[start:end])}</mark>'
        )
        cursor = end
    parts.append(_escape(raw[cursor:]))
    return "".join(parts)


def _option_cards(record):
    options = record.get("parsed_options", {}) or {}
    selected = labels_by_id.get(str(record["id"]))
    heuristic = record.get("heuristic_answer")
    cards = []
    for letter in ["A", "B", "C", "D"]:
        option = options.get(letter, "")
        is_selected = selected == letter
        is_heuristic = heuristic == letter
        border = "2px solid #2563eb" if is_selected else "1px solid #d0d7de"
        bg = "#eff6ff" if is_selected else "#ffffff"
        tags = []
        if is_selected:
            tags.append('<span style="color:#2563eb; font-weight:600;">labelled</span>')
        if is_heuristic:
            tags.append('<span style="color:#7c2d12; font-weight:600;">heuristic</span>')
        tag_html = " ┬и ".join(tags)
        cards.append(
            f'<div style="border:{border}; background:{bg}; border-radius:6px; padding:9px 11px; margin:7px 0;">'
            f'<div style="display:flex; justify-content:space-between; gap:12px;">'
            f'<b>{letter}.</b><span>{tag_html}</span></div>'
            f'<div style="margin-top:4px; white-space:pre-wrap;">{_escape(option)}</div>'
            f'</div>'
        )
    return "".join(cards)


def _render():
    record = _current_record()
    row_id = str(record["id"])
    current = labels_by_id.get(row_id, "")
    heuristic = record.get("heuristic_answer") or "None"
    heuristic_source = record.get("heuristic_answer_source") or ""
    parse_error = record.get("parse_error") or ""
    options = record.get("parsed_options", {}) or {}

    desired_jump = label_state["index"] + 1
    if jump_box.value != desired_jump:
        jump_box.value = desired_jump

    labelled = sum(1 for rec in korean_hard_records if labels_by_id.get(str(rec["id"])))
    with status_out:
        clear_output(wait=True)
        print(f"Labelled {labelled}/{len(korean_hard_records)} | Current ID {row_id} | Output: {LABEL_OUTPUT_PATH}")

    parse_error_html = ""
    if parse_error:
        parse_error_html = f'<div style="color:#b42318; margin:8px 0;"><b>Parse error:</b> {_escape(parse_error)}</div>'

    chosen_html = _highlight_exact_option_matches(record.get("chosen", ""), options)
    rejected_html = _highlight_exact_option_matches(record.get("rejected", ""), options)
    prompt_html = _escape(record.get("prompt", ""))

    html_block = f'''
    <div style="font-family: system-ui, -apple-system, Segoe UI, sans-serif; line-height:1.45;">
      <div style="display:flex; flex-wrap:wrap; gap:14px; align-items:center; margin:8px 0 12px;">
        <span><b>Subset:</b> {_escape(record.get("subset", ""))}</span>
        <span><b>ID:</b> {_escape(row_id)}</span>
        <span><b>Label:</b> {_escape(current or "unlabelled")}</span>
        <span><b>Heuristic:</b> {_escape(heuristic)} {_escape(f"({heuristic_source})" if heuristic_source else "")}</span>
      </div>
      {parse_error_html}
      <h4 style="margin-bottom:6px;">Parsed Question</h4>
      <div style="white-space:pre-wrap; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#f6f8fa;">{_escape(record.get("parsed_question", ""))}</div>
      <h4 style="margin-bottom:6px;">Parsed Answers</h4>
      {_option_cards(record)}
      <h4 style="margin-bottom:6px;">Prompt</h4>
      <div style="white-space:pre-wrap; max-height:230px; overflow:auto; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#fff;">{prompt_html}</div>
      <h4 style="margin-bottom:6px;">Chosen / Reference <span style="font-size:12px; font-weight:400;">yellow = exact answer option text match</span></h4>
      <div style="white-space:pre-wrap; max-height:300px; overflow:auto; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#fff;">{chosen_html}</div>
      <h4 style="margin-bottom:6px;">Rejected <span style="font-size:12px; font-weight:400;">yellow = exact answer option text match</span></h4>
      <div style="white-space:pre-wrap; max-height:300px; overflow:auto; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#fff;">{rejected_html}</div>
    </div>
    '''

    with item_out:
        clear_output(wait=True)
        display(HTML(html_block))


def _persist():
    save_korean_hard_labels(labels_by_id)


def _set_label(answer):
    labels_by_id[str(_current_record()["id"])] = answer
    _persist()
    if label_state["index"] < len(korean_hard_records) - 1:
        label_state["index"] += 1
    _render()


def _clear_label(_=None):
    labels_by_id.pop(str(_current_record()["id"]), None)
    _persist()
    _render()


def _save(_=None):
    _persist()
    _render()


def _prev(_=None):
    label_state["index"] = max(0, label_state["index"] - 1)
    _render()


def _next(_=None):
    label_state["index"] = min(len(korean_hard_records) - 1, label_state["index"] + 1)
    _render()


def _jump(change):
    label_state["index"] = int(change["new"]) - 1
    _render()

for button in answer_buttons:
    button.on_click(lambda _, letter=button.description: _set_label(letter))
unknown_button.on_click(lambda _: _set_label("UNKNOWN"))
clear_button.on_click(_clear_label)
save_button.on_click(_save)
prev_button.on_click(_prev)
next_button.on_click(_next)
jump_box.observe(_jump, names="value")

controls = widgets.HBox(answer_buttons + [unknown_button, clear_button, prev_button, next_button, jump_box, save_button])
display(widgets.VBox([controls, status_out, item_out]))
_render()
```

## Backfill Rejected Responses

Run this if `rejected` is missing or empty in `kudge_korean_hard_groundtruth_labels.json`. It reloads `amphora/kudge-challenge`, copies `rejected`, `chosen_model`, and `rejected_model` into the existing label JSON, and preserves your `groundtruth_answer` labels.


```python
from datasets import load_dataset


def backfill_rejected_from_kudge_challenge(path=LABEL_OUTPUT_PATH):
    records = json.loads(path.read_text(encoding="utf-8"))
    ds = load_dataset("amphora/kudge-challenge", split="train")
    raw_by_id = {
        str(row["id"]): row
        for row in ds
        if row.get("subset") == "Korean-Hard"
    }

    updated = 0
    missing = []
    for record in records:
        row = raw_by_id.get(str(record.get("id")))
        if row is None:
            missing.append(record.get("id"))
            continue
        before = (record.get("rejected", ""), record.get("chosen_model", ""), record.get("rejected_model", ""))
        record["prompt"] = row.get("prompt", record.get("prompt", ""))
        record["chosen"] = row.get("chosen", record.get("chosen", ""))
        record["rejected"] = row.get("rejected", record.get("rejected", ""))
        record["chosen_model"] = row.get("chosen_model", record.get("chosen_model", ""))
        record["rejected_model"] = row.get("rejected_model", record.get("rejected_model", ""))
        after = (record.get("rejected", ""), record.get("chosen_model", ""), record.get("rejected_model", ""))
        if after != before:
            updated += 1

    path.write_text(json.dumps(records, ensure_ascii=False, indent=2), encoding="utf-8")
    rejected_count = sum(bool(str(record.get("rejected", "")).strip()) for record in records)
    print(f"Updated {updated}/{len(records)} records")
    print(f"Rejected present: {rejected_count}/{len(records)}")
    if missing:
        print(f"Missing IDs: {missing}")
    return records

# Run when needed:
# korean_hard_records = backfill_rejected_from_kudge_challenge()
```

## Verify Rejected Responses

Reload Korean-Hard items from `amphora/kudge-challenge` and compare the source `rejected` text against the `rejected` stored in `kudge_korean_hard_groundtruth_labels.json`. If the mismatch table is empty, the JSON is synchronized.


```python
from datasets import load_dataset


def verify_rejected_matches_source(path=LABEL_OUTPUT_PATH):
    stored_records = json.loads(path.read_text(encoding="utf-8"))
    stored_by_id = {str(record.get("id")): record for record in stored_records}

    source_ds = load_dataset("amphora/kudge-challenge", split="train")
    source_rows = [row for row in source_ds if row.get("subset") == "Korean-Hard"]
    source_rows.sort(key=lambda row: int(row["id"]))

    rows = []
    for source in source_rows:
        row_id = str(source["id"])
        stored = stored_by_id.get(row_id, {})
        source_rejected = source.get("rejected", "") or ""
        stored_rejected = stored.get("rejected", "") or ""
        rows.append({
            "id": row_id,
            "source_has_rejected": bool(source_rejected.strip()),
            "stored_has_rejected": bool(stored_rejected.strip()),
            "rejected_matches": source_rejected == stored_rejected,
            "source_rejected_len": len(source_rejected),
            "stored_rejected_len": len(stored_rejected),
            "source_rejected_preview": source_rejected[:220],
            "stored_rejected_preview": stored_rejected[:220],
            "source_rejected_model": source.get("rejected_model", ""),
            "stored_rejected_model": stored.get("rejected_model", ""),
            "rejected_model_matches": source.get("rejected_model", "") == stored.get("rejected_model", ""),
        })

    verify_df = pd.DataFrame(rows)
    print(f"Source Korean-Hard rows: {len(source_rows)}")
    print(f"Stored JSON rows: {len(stored_records)}")
    print(f"Rejected matches: {int(verify_df['rejected_matches'].sum())}/{len(verify_df)}")
    print(f"Rejected model matches: {int(verify_df['rejected_model_matches'].sum())}/{len(verify_df)}")
    return verify_df


rejected_verify_df = verify_rejected_matches_source()
rejected_mismatch_df = rejected_verify_df[
    (~rejected_verify_df["rejected_matches"])
    | (~rejected_verify_df["rejected_model_matches"])
].copy()
rejected_mismatch_df
```

## Review UNKNOWN Labels

Use this smaller interface after your first pass. It loads the saved Korean-Hard JSON, filters items whose `groundtruth_answer` is `UNKNOWN`, and lets you replace them with `A/B/C/D` after researching. It writes back to the same JSON file.


```python
from pathlib import Path
import json
import pandas as pd

try:
    UNKNOWN_REVIEW_PATH = LABEL_OUTPUT_PATH
except NameError:
    UNKNOWN_REVIEW_PATH = repo_root / "benchmarks" / "kudge" / "kudge_korean_hard_groundtruth_labels.json"


def load_label_records(path=UNKNOWN_REVIEW_PATH):
    with path.open("r", encoding="utf-8") as fh:
        return json.load(fh)


def save_label_records(records, path=UNKNOWN_REVIEW_PATH):
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open("w", encoding="utf-8") as fh:
        json.dump(records, fh, ensure_ascii=False, indent=2)
    print(f"Saved {len(records)} records to {path}")


def refresh_unknown_review_records(include_blank=False):
    all_records = load_label_records()
    review_records = [
        record for record in all_records
        if record.get("groundtruth_answer") == "UNKNOWN"
        or (include_blank and not record.get("groundtruth_answer"))
    ]
    return all_records, review_records

all_label_records, unknown_review_records = refresh_unknown_review_records(include_blank=False)
print(f"UNKNOWN records to review: {len(unknown_review_records)}")
if unknown_review_records:
    display(pd.DataFrame(unknown_review_records)[["id", "subset", "groundtruth_answer", "heuristic_answer", "heuristic_answer_source"]].head())
else:
    print("No UNKNOWN labels remain.")
```

```python
import html
import re
import ipywidgets as widgets
from IPython.display import HTML, display, clear_output

unknown_state = {"index": 0}
unknown_out = widgets.Output()
unknown_status_out = widgets.Output()

unknown_answer_buttons = [
    widgets.Button(description=letter, button_style="info", layout=widgets.Layout(width="64px", height="36px"))
    for letter in ["A", "B", "C", "D"]
]
unknown_keep_button = widgets.Button(description="Keep UNKNOWN", icon="help-circle", layout=widgets.Layout(width="145px", height="36px"))
unknown_prev_button = widgets.Button(description="Previous", icon="arrow-left", layout=widgets.Layout(width="115px", height="36px"))
unknown_next_button = widgets.Button(description="Next", icon="arrow-right", layout=widgets.Layout(width="105px", height="36px"))
unknown_save_button = widgets.Button(description="Save JSON", icon="save", button_style="success", layout=widgets.Layout(width="125px", height="36px"))
unknown_refresh_button = widgets.Button(description="Refresh", icon="refresh-cw", layout=widgets.Layout(width="105px", height="36px"))
unknown_jump_box = widgets.BoundedIntText(value=1, min=1, max=max(len(unknown_review_records), 1), description="Unknown", layout=widgets.Layout(width="190px"))


def _u_escape(value):
    return html.escape(str(value))


def _u_highlight_exact_option_matches(text, options):
    raw = str(text)
    matches = []
    for letter, option in (options or {}).items():
        option = str(option).strip()
        if len(option) < 2:
            continue
        for match in re.finditer(re.escape(option), raw, flags=re.IGNORECASE):
            matches.append((match.start(), match.end(), str(letter)))
    if not matches:
        return _u_escape(raw)

    matches.sort(key=lambda x: (x[0], -(x[1] - x[0])))
    merged = []
    last_end = -1
    for start, end, letter in matches:
        if start < last_end:
            continue
        merged.append((start, end, letter))
        last_end = end

    parts = []
    cursor = 0
    for start, end, letter in merged:
        parts.append(_u_escape(raw[cursor:start]))
        parts.append(
            '<mark style="background:#fff59d; padding:1px 3px; border-radius:3px;" '
            f'title="exact match for option {letter}">{_u_escape(raw[start:end])}</mark>'
        )
        cursor = end
    parts.append(_u_escape(raw[cursor:]))
    return "".join(parts)


def _reload_unknowns(keep_index=True):
    global all_label_records, unknown_review_records
    old_index = unknown_state["index"]
    all_label_records, unknown_review_records = refresh_unknown_review_records(include_blank=False)
    unknown_jump_box.max = max(len(unknown_review_records), 1)
    if keep_index:
        unknown_state["index"] = min(old_index, max(len(unknown_review_records) - 1, 0))
    else:
        unknown_state["index"] = 0


def _unknown_current_record():
    if not unknown_review_records:
        return None
    unknown_state["index"] = max(0, min(unknown_state["index"], len(unknown_review_records) - 1))
    return unknown_review_records[unknown_state["index"]]


def _set_record_label(record_id, answer):
    updated = False
    for record in all_label_records:
        if str(record.get("id")) == str(record_id):
            record["groundtruth_answer"] = answer
            updated = True
            break
    if not updated:
        raise KeyError(f"Could not find record id {record_id}")
    save_label_records(all_label_records)


def _unknown_option_cards(record):
    options = record.get("parsed_options", {}) or {}
    selected = record.get("groundtruth_answer")
    cards = []
    for letter in ["A", "B", "C", "D"]:
        option = options.get(letter, "")
        is_selected = selected == letter
        border = "2px solid #2563eb" if is_selected else "1px solid #d0d7de"
        bg = "#eff6ff" if is_selected else "#ffffff"
        cards.append(
            f'<div style="border:{border}; background:{bg}; border-radius:6px; padding:9px 11px; margin:7px 0;">'
            f'<b>{letter}.</b> <span style="white-space:pre-wrap;">{_u_escape(option)}</span>'
            f'</div>'
        )
    return "".join(cards)


def _unknown_render():
    record = _unknown_current_record()
    with unknown_status_out:
        clear_output(wait=True)
        remaining = sum(1 for record in all_label_records if record.get("groundtruth_answer") == "UNKNOWN")
        print(f"UNKNOWN remaining in file: {remaining} | Review list size: {len(unknown_review_records)} | Output: {UNKNOWN_REVIEW_PATH}")

    with unknown_out:
        clear_output(wait=True)
        if record is None:
            display(HTML('<div style="font-family:system-ui; padding:12px; border:1px solid #d0d7de; border-radius:6px;">No UNKNOWN records to review.</div>'))
            return

        desired = unknown_state["index"] + 1
        if unknown_jump_box.value != desired:
            unknown_jump_box.value = desired

        options = record.get("parsed_options", {}) or {}
        chosen_html = _u_highlight_exact_option_matches(record.get("chosen", ""), options)
        heuristic = record.get("heuristic_answer") or "None"
        heuristic_source = record.get("heuristic_answer_source") or ""
        html_block = f'''
        <div style="font-family: system-ui, -apple-system, Segoe UI, sans-serif; line-height:1.45;">
          <div style="display:flex; flex-wrap:wrap; gap:14px; align-items:center; margin:8px 0 12px;">
            <span><b>UNKNOWN item:</b> {unknown_state["index"] + 1}/{len(unknown_review_records)}</span>
            <span><b>Subset:</b> {_u_escape(record.get("subset", ""))}</span>
            <span><b>ID:</b> {_u_escape(record.get("id", ""))}</span>
            <span><b>Current:</b> {_u_escape(record.get("groundtruth_answer", ""))}</span>
            <span><b>Heuristic:</b> {_u_escape(heuristic)} {_u_escape(f"({heuristic_source})" if heuristic_source else "")}</span>
          </div>
          <h4 style="margin-bottom:6px;">Parsed Question</h4>
          <div style="white-space:pre-wrap; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#f6f8fa;">{_u_escape(record.get("parsed_question", ""))}</div>
          <h4 style="margin-bottom:6px;">Parsed Answers</h4>
          {_unknown_option_cards(record)}
          <h4 style="margin-bottom:6px;">Prompt</h4>
          <div style="white-space:pre-wrap; max-height:230px; overflow:auto; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#fff;">{_u_escape(record.get("prompt", ""))}</div>
          <h4 style="margin-bottom:6px;">Chosen / Reference <span style="font-size:12px; font-weight:400;">yellow = exact answer option text match</span></h4>
          <div style="white-space:pre-wrap; max-height:340px; overflow:auto; border:1px solid #d0d7de; border-radius:6px; padding:10px; background:#fff;">{chosen_html}</div>
        </div>
        '''
        display(HTML(html_block))


def _unknown_set_label(answer):
    record = _unknown_current_record()
    if record is None:
        return
    _set_record_label(record["id"], answer)
    _reload_unknowns(keep_index=True)
    _unknown_render()


def _unknown_keep(_=None):
    if unknown_review_records and unknown_state["index"] < len(unknown_review_records) - 1:
        unknown_state["index"] += 1
    _unknown_render()


def _unknown_prev(_=None):
    unknown_state["index"] = max(0, unknown_state["index"] - 1)
    _unknown_render()


def _unknown_next(_=None):
    unknown_state["index"] = min(max(len(unknown_review_records) - 1, 0), unknown_state["index"] + 1)
    _unknown_render()


def _unknown_save(_=None):
    save_label_records(all_label_records)
    _reload_unknowns(keep_index=True)
    _unknown_render()


def _unknown_refresh(_=None):
    _reload_unknowns(keep_index=False)
    _unknown_render()


def _unknown_jump(change):
    unknown_state["index"] = int(change["new"]) - 1
    _unknown_render()

for button in unknown_answer_buttons:
    button.on_click(lambda _, letter=button.description: _unknown_set_label(letter))
unknown_keep_button.on_click(_unknown_keep)
unknown_prev_button.on_click(_unknown_prev)
unknown_next_button.on_click(_unknown_next)
unknown_save_button.on_click(_unknown_save)
unknown_refresh_button.on_click(_unknown_refresh)
unknown_jump_box.observe(_unknown_jump, names="value")

unknown_controls = widgets.HBox(unknown_answer_buttons + [unknown_keep_button, unknown_prev_button, unknown_next_button, unknown_jump_box, unknown_save_button, unknown_refresh_button])
display(widgets.VBox([unknown_controls, unknown_status_out, unknown_out]))
_unknown_render()
```

## Qwen 4B Answer Extraction

These cells use a local Hugging Face Qwen 4B model to infer answer letters from the reference solution text when the heuristic extractor cannot find one. Run the install cell first, then load the model, then start with `limit=5`.


```python
# Run once if Qwen dependencies are not installed in this kernel.
# On this x86_64 macOS/Python setup, torch tops out at 2.2.2,
# so keep transformers on a compatible 4.x release.
%pip install -U "numpy<2" "torch==2.2.2" "transformers==4.51.3" accelerate
```

```python
import json
import time
from pathlib import Path

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

QWEN_MODEL_ID = "Qwen/Qwen3-4B-Instruct-2507"

tokenizer = AutoTokenizer.from_pretrained(QWEN_MODEL_ID, trust_remote_code=True)
qwen_model = AutoModelForCausalLM.from_pretrained(
    QWEN_MODEL_ID,
    torch_dtype="auto",
    device_map="auto",
    trust_remote_code=True,
)
qwen_model.eval()
```

```python
QWEN_CACHE_PATH = repo_root / "benchmarks" / "results" / "kudge_qwen4b_answer_cache.jsonl"

def qwen_answer_prompt(row):
    parsed = parse_mcq(row["prompt"])
    return f"""You are matching a reference solution to a multiple-choice answer.
Return only compact JSON with keys answer and rationale.
answer must be one of A, B, C, D, or null if the reference does not determine the answer.

Question:
{parsed["question"]}

Options:
A. {parsed["options"].get("A", "")}
B. {parsed["options"].get("B", "")}
C. {parsed["options"].get("C", "")}
D. {parsed["options"].get("D", "")}

Reference solution:
{row["chosen"]}

JSON:"""

def _qwen_json_from_text(text):
    match = re.search(r"\{.*\}", str(text), flags=re.DOTALL)
    if not match:
        return {"answer": None, "rationale": str(text).strip()}
    try:
        data = json.loads(match.group(0))
    except json.JSONDecodeError:
        return {"answer": None, "rationale": str(text).strip()}
    answer = data.get("answer")
    if isinstance(answer, str):
        answer = answer.strip().upper()
    if answer not in {"A", "B", "C", "D"}:
        answer = None
    return {"answer": answer, "rationale": str(data.get("rationale", "")).strip()}

def ask_qwen_answer(row, max_new_tokens=96):
    messages = [
        {"role": "system", "content": "You identify the correct multiple-choice option from a reference solution. Return JSON only."},
        {"role": "user", "content": qwen_answer_prompt(row)},
    ]
    try:
        prompt = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True,
            enable_thinking=False,
        )
    except TypeError:
        prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    device = next(qwen_model.parameters()).device
    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {key: value.to(device) for key, value in inputs.items()}

    with torch.no_grad():
        generated = qwen_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
        )
    new_tokens = generated[0][inputs["input_ids"].shape[-1]:]
    raw = tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
    parsed = _qwen_json_from_text(raw)
    parsed["raw"] = raw
    return parsed

def load_qwen_cache(path=QWEN_CACHE_PATH):
    if not path.exists():
        return {}
    cache = {}
    with path.open("r", encoding="utf-8") as fh:
        for line in fh:
            if line.strip():
                record = json.loads(line)
                cache[str(record["id"])] = record
    return cache

def append_qwen_cache(record, path=QWEN_CACHE_PATH):
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open("a", encoding="utf-8") as fh:
        fh.write(json.dumps(record, ensure_ascii=False) + "\n")

def run_qwen_answer_extraction(rows_df, *, only_missing=True, limit=None, sleep_s=0):
    cache = load_qwen_cache()
    records = []
    attempted = 0

    for idx, row in rows_df.iterrows():
        row_id = str(row["id"])
        if only_missing and "mcq_df" in globals():
            existing = mcq_df.loc[idx, "answer"] if idx in mcq_df.index else None
            if pd.notna(existing) and existing:
                continue

        if row_id in cache:
            records.append(cache[row_id])
            continue

        if limit is not None and attempted >= limit:
            break
        attempted += 1

        result = ask_qwen_answer(row)
        parsed = parse_mcq(row["prompt"])
        record = {
            "id": row_id,
            "subset": row["subset"],
            "answer": result["answer"],
            "answer_text": parsed["options"].get(result["answer"], "") if result["answer"] else "",
            "rationale": result["rationale"],
            "raw": result["raw"],
            "model": QWEN_MODEL_ID,
        }
        append_qwen_cache(record)
        records.append(record)
        if sleep_s:
            time.sleep(sleep_s)

    return pd.DataFrame(records)
```

## Check Labels Against GPQA Main CSV

`gpqa_main.csv` is not row-aligned with the Korean-Hard labels, but the source order appears to be preserved. This section uses monotonic matching across GPQA rows, then checks whether your labeled KUDGE option text matches GPQA's English `Correct Answer`. Rows with translated chemical names or weak text overlap are kept for manual review.


```python
from pathlib import Path
import csv
import json
import re
import unicodedata
from collections import Counter

import pandas as pd

LABELS_JSON_PATH = repo_root / "benchmarks" / "kudge" / "kudge_korean_hard_groundtruth_labels.json"
GPQA_MAIN_CSV_PATH = repo_root / "benchmarks" / "kudge" / "gpqa_main.csv"
GPQA_MATCH_REPORT_PATH = repo_root / "benchmarks" / "kudge" / "checking(ignore)" / "kudge_gpqa_match_review.csv"


def _clean_text(value):
    value = "" if value is None else str(value)
    value = unicodedata.normalize("NFKC", value)
    value = value.replace("РёЈ", "hbar").replace("¤ђ", "pi").replace("├Ќ", "x").replace("Рѕњ", "-")
    return re.sub(r"\s+", " ", value).strip().lower()


def _compact_english(value):
    return re.sub(r"[^a-z0-9]+", "", _clean_text(value))


def _tokens(value):
    value = _clean_text(value)
    toks = re.findall(r"[a-z]+\d*|\d+(?:\.\d+)?(?:e[+-]?\d+)?|[a-z0-9]+", value)
    stop = {"the", "and", "for", "with", "that", "this", "which", "what", "from", "are", "is", "of", "to", "in", "an", "by", "as", "be", "it", "or", "on", "one", "two"}
    return {tok for tok in toks if len(tok) > 1 and tok not in stop}


def _jaccard(left, right):
    return len(left & right) / len(left | right) if left and right else 0.0


def _option_score(kudge_option, gpqa_answer):
    left = _compact_english(kudge_option)
    right = _compact_english(gpqa_answer)
    if not left or not right:
        return 0.0
    if left == right:
        return 1.0
    if len(left) >= 2 and left in right:
        return 0.95
    if len(right) >= 2 and right in left:
        return 0.95
    return _jaccard(_tokens(kudge_option), _tokens(gpqa_answer))


def load_gpqa_match_inputs(labels_path=LABELS_JSON_PATH, gpqa_path=GPQA_MAIN_CSV_PATH):
    labels = json.loads(Path(labels_path).read_text(encoding="utf-8"))
    labels = [row for row in labels if row.get("subset") == "Korean-Hard"]
    gpqa = pd.read_csv(gpqa_path).fillna("")
    required = ["Question", "Correct Answer", "Incorrect Answer 1", "Incorrect Answer 2", "Incorrect Answer 3"]
    missing = [col for col in required if col not in gpqa.columns]
    if missing:
        raise ValueError(f"Missing expected GPQA columns: {missing}")
    return labels, gpqa


def _best_option_alignment(parsed_options, gpqa_row):
    answer_cols = {
        "correct": "Correct Answer",
        "incorrect_1": "Incorrect Answer 1",
        "incorrect_2": "Incorrect Answer 2",
        "incorrect_3": "Incorrect Answer 3",
    }
    best_letter, best_kind, best_score = "", "", 0.0
    for letter, option in parsed_options.items():
        for kind, col in answer_cols.items():
            score = _option_score(option, gpqa_row[col])
            if score > best_score:
                best_letter, best_kind, best_score = letter, kind, score
    return best_letter, best_kind, best_score



def _monotonic_assign(score_matrix):
    """Return increasing GPQA row assignments that maximize total match score."""
    n = len(score_matrix)
    m = len(score_matrix[0]) if n else 0
    neg = -10**9
    dp = [[neg] * (m + 1) for _ in range(n + 1)]
    keep = [[False] * (m + 1) for _ in range(n + 1)]
    for col in range(m + 1):
        dp[0][col] = 0.0
    for row in range(1, n + 1):
        for col in range(1, m + 1):
            skip = dp[row][col - 1]
            take = dp[row - 1][col - 1] + score_matrix[row - 1][col - 1]
            if take > skip:
                dp[row][col] = take
                keep[row][col] = True
            else:
                dp[row][col] = skip

    assignments = [None] * n
    row, col = n, m
    while row > 0 and col > 0:
        if keep[row][col]:
            assignments[row - 1] = col - 1
            row -= 1
            col -= 1
        else:
            col -= 1
    return assignments


def build_gpqa_match_report(labels, gpqa_df):
    gpqa_prepped = []
    for gpqa_idx, gpqa_row in gpqa_df.iterrows():
        answer_text = " ".join(str(gpqa_row[col]) for col in ["Correct Answer", "Incorrect Answer 1", "Incorrect Answer 2", "Incorrect Answer 3"])
        all_text = " ".join([str(gpqa_row["Question"]), answer_text, str(gpqa_row.get("Explanation", ""))])
        gpqa_prepped.append((gpqa_idx, gpqa_row, _tokens(all_text), _tokens(answer_text)))

    score_matrix = []
    meta_matrix = []
    for item in labels:
        parsed_options = item.get("parsed_options") or {}
        groundtruth_answer = str(item.get("groundtruth_answer", "")).strip().upper()
        groundtruth_text = parsed_options.get(groundtruth_answer, "")
        kudge_all_tokens = _tokens(" ".join([
            item.get("prompt", ""),
            item.get("chosen", ""),
            item.get("parsed_question", ""),
            " ".join(parsed_options.values()),
        ]))
        kudge_option_tokens = _tokens(" ".join(parsed_options.values()))

        item_scores = []
        item_meta = []
        for gpqa_idx, gpqa_row, gpqa_all_tokens, gpqa_answer_tokens in gpqa_prepped:
            best_letter, best_kind, best_option_score = _best_option_alignment(parsed_options, gpqa_row)
            direct_correct_score = _option_score(groundtruth_text, gpqa_row["Correct Answer"])
            match_score = (
                0.50 * best_option_score
                + 0.30 * _jaccard(kudge_all_tokens, gpqa_all_tokens)
                + 0.20 * _jaccard(kudge_option_tokens, gpqa_answer_tokens)
            )
            item_scores.append(match_score)
            item_meta.append((direct_correct_score, best_letter, best_kind, best_option_score))
        score_matrix.append(item_scores)
        meta_matrix.append(item_meta)

    assignments = _monotonic_assign(score_matrix)
    report_rows = []
    for item_idx, (item, assigned_gpqa_pos) in enumerate(zip(labels, assignments)):
        parsed_options = item.get("parsed_options") or {}
        groundtruth_answer = str(item.get("groundtruth_answer", "")).strip().upper()
        groundtruth_text = parsed_options.get(groundtruth_answer, "")
        gpqa_idx, gpqa_row, _, _ = gpqa_prepped[assigned_gpqa_pos]
        match_score = score_matrix[item_idx][assigned_gpqa_pos]
        independent_best_score = max(score_matrix[item_idx])
        direct_correct_score, best_letter, best_kind, best_option_score = meta_matrix[item_idx][assigned_gpqa_pos]
        direct_match = direct_correct_score >= 0.88
        confidence = "high" if match_score >= 0.30 and direct_match else "medium" if match_score >= 0.16 else "low"
        verdict = "LABEL_MATCHES_GPQA_CORRECT" if direct_match else "NEEDS_MANUAL_REVIEW"

        report_rows.append({
            "kudge_id": item.get("id", ""),
            "subset": item.get("subset", ""),
            "groundtruth_answer": groundtruth_answer,
            "groundtruth_answer_text": groundtruth_text,
            "matched_gpqa_row": int(gpqa_idx),
            "match_score": round(match_score, 3),
            "independent_best_score": round(independent_best_score, 3),
            "match_confidence": confidence,
            "gpqa_record_id": gpqa_row.get("Record ID", ""),
            "gpqa_question": gpqa_row["Question"],
            "gpqa_correct_answer": gpqa_row["Correct Answer"],
            "gpqa_incorrect_1": gpqa_row["Incorrect Answer 1"],
            "gpqa_incorrect_2": gpqa_row["Incorrect Answer 2"],
            "gpqa_incorrect_3": gpqa_row["Incorrect Answer 3"],
            "groundtruth_vs_gpqa_correct_score": round(direct_correct_score, 3),
            "best_option_alignment": f"{best_letter}->{best_kind} ({best_option_score:.3f})",
            "verdict": verdict,
        })
    return pd.DataFrame(report_rows)

korean_hard_labels, gpqa_main_df = load_gpqa_match_inputs()
gpqa_match_report_df = build_gpqa_match_report(korean_hard_labels, gpqa_main_df)
print(f"Labels: {len(korean_hard_labels)} | GPQA rows: {len(gpqa_main_df)}")
print(Counter(gpqa_match_report_df["verdict"]))
print(Counter(gpqa_match_report_df["match_confidence"]))
gpqa_match_report_df.head()
```
